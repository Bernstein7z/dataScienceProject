```python
import json
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import holoviews as hv
from holoviews import opts

matplotlib.rcParams['axes.titlesize'] = 30
matplotlib.rcParams['axes.titlepad'] = 30

hv.extension('bokeh')
```

```python
def get_sections():
    with open('../data/authors.json') as json_data:
        data = json.load(json_data)

        sections = {}
        for values in data.values():
            for key in values.keys():
                if key != 'co_authors':
                    if key not in sections:
                        sections[key] = 1
                    else:
                        sections[key] += 1

        return sections
```

```python
def get_opinions():
    with open('../data/authors.json') as json_data:
        data = json.load(json_data)

        tmp = []
        for authors in data.values():
            if 'OpEd' in authors and len(authors.keys()) > 1:
                for key, value in authors.items():
                    if key != 'co_authors' and key != 'OpEd':
                        tmp.append(['OpEd', key, value])
        
        return tmp
        
```

```python
def get_name(person: dict) -> str:
    result = ''
    
    if 'firstname' in person and person['firstname'] is not None: result += person['firstname'] 
    if 'middlename' in person and person['middlename'] is not None: result += person['middlename']
    if 'lastname' in person and person['lastname'] is not None: result += person['lastname']

    return result.lower()
    
```

```python
def get_data():
    words = []
    authors = []
    articles = []
    persons = {}

    for iy, year in enumerate(range(2016, 2021)):
        print(year, sep=' ', end=' ')
        for im, month in enumerate(range(1, 13)):
            with open(f'../data/{year}_{month}.json') as json_file:
                data = json.load(json_file)

                word = 0
                author = []
                article = 0
                
                for doc in data['response']['docs']:    
                    if doc['document_type'] == 'article':
                        article += 1
                        if 'word_count' in doc: word += doc['word_count']
                        
                    if 'byline' in doc and 'person' in doc['byline']:
                        if doc['byline']['person'] is not None and len(doc['byline']['person']) > 0:
                            for person in doc['byline']['person']:
                                author_name = get_name(person)
                                author.append(author_name)

                                if author_name in persons:
                                    persons[author_name]['last'] = (iy+1) * (im+1)
                                else:
                                    persons[author_name] = {
                                        'first': (iy+1) * (im+1),
                                        'last': None 
                                    }
                
                words.append(word)
                authors.append(len(set(author)))
                articles.append(article)
    
    return words, authors, articles, persons
```

```python
sections = get_sections()
sosec = sorted(sections.items(), key=lambda y: y[1], reverse=True)
sorted_sections = list(filter(lambda x: x[1] > 10, sosec))

values = [x[1] for x in sorted_sections]
labels = [x[0] for x in sorted_sections]

plt.figure(1, figsize=(25.0, 10.0))
plt.bar(labels, height=values, align='center')

plt.title('Authors (> 10 articles) Distribution')
plt.grid(axis='y')
plt.xlim(-0.5)
plt.xticks(rotation='vertical')
plt.ylabel('Articles', fontsize=20, labelpad=20)

plt.show()
```

```python
tmp = get_opinions()
opEd = pd.DataFrame(tmp, columns=['source', 'target', 'values'])
opEd['total'] = opEd.groupby(['target'])['values'].transform('sum')
opEd = opEd.drop_duplicates(subset=['target'])
del opEd['values']
opEd = opEd.rename(columns={'total': 'value'}).reset_index(drop=True)
opEd.astype({'value': 'int32'})
opEd = opEd.loc[opEd['value'] > 1000]

image = hv.Sankey(opEd, label='OpEd').opts(
    opts.Sankey(label_position='left', edge_color='target', node_padding=10, node_color='index', cmap='tab20', width=1500, height=900))
hv.save(image, 'OpEd.html', backend='bokeh')
```

```python
words, authors, articles, persons = get_data()
```
